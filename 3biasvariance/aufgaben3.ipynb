### 1. Gehaltsunterschied Männer / Frauen

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
salaries = pd.read_csv('salaries.csv')
salaries.head()
```

```python
# linear regression model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# features is only sex
X = salaries.copy()['sex'].values.reshape(-1,1)
y = salaries.copy()['salary'].values.reshape(-1,1)

# features are 'male' or 'female'
# should transform to numeric values
male = 0
female = 1
for i, sex in enumerate(X):
    X[i] = male if sex == 'Male' else female

X.shape, y.shape
```

```python
# split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

X_train, y_train
```

```python
reg = LinearRegression().fit(X_train, y_train)

print('a0 (mens average salary):', reg.intercept_[0])
print('a1 (women salary difference):', reg.coef_[0][0])
```

Der Intercept stellt den durschschnittlichen Gehalt der Männer da (feature ist 0). a1 stellt die Differenz im Gehalt dar. Frauen verdienen dementsprechend weniger.

### 2. Lineare Regression mit allen Variablen

```python
salaries.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
X = salaries.drop(['salary'], axis=1)
y = salaries['salary'].values.reshape(-1,1)
X.loc[0]
```

```python
# transform to numeric values: rank, discipline, sex
enc = OneHotEncoder(drop='first')
enc.fit(X)
enc.categories_
```

```python
X = enc.transform(X).toarray()
```

```python
reg = LinearRegression().fit(X, y)
```

```python
error = np.mean((reg.predict(X) - y)**2)
error
```

Ergebnisse

```python
# interpret coefficients
print('a0 (womans assoc. prof discipline A average salary):', reg.intercept_[0])

for i, col in enumerate(enc.categories_):
    # exclude numeric values
    if all(isinstance(x, np.int64) for x in col):
        continue
    
    for j in range(1, len(col)):
        print('a{} ({}): {}'.format(i+j, col[j], reg.coef_[0][i+j-1]))
```
