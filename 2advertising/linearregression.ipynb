## Lineare Regression: `sales ~ TV`

```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
```

```python
X = advertising['TV'].values.reshape(-1,1)
y = advertising['sales'].values.reshape(-1,1)
```

```python
a0_list = []
a1_list = []
train_mqa_list = []
test_mqa_list = []

for i in range(2000):
    print(f"Versuch {i+1}")
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)

    regressor = LinearRegression()
    regressor.fit(X_train, y_train)
    print("    Koeffizienten der Regressiongerade: ")
    print(f"        a0: {regressor.intercept_[0]}")
    print(f"        a1: {regressor.coef_[0][0]}")

    # error
    train_mqa = np.mean((y_train - regressor.predict(X_train)) ** 2)
    print(f"    Train MQA: {train_mqa}")
    test_mqa = np.mean((y_test - regressor.predict(X_test)) ** 2)
    print(f"    Test MQA: {test_mqa}")

    a0_list.append(regressor.intercept_[0])
    a1_list.append(regressor.coef_[0][0])
    train_mqa_list.append(train_mqa)
    test_mqa_list.append(test_mqa)

    # Plotten
    # plt.xlabel('TV ($)')
    # plt.ylabel('Sales ($)')
    # plt.scatter(X_train, y_train, color='gray')
    # plt.scatter(X_test, y_test, color='blue')
    # plt.legend(['Train', 'Test'])
    # plt.plot(X_train, regressor.predict(X_train), color='red', linewidth=2)
    # plt.show()
```

```python
plt.hist(a0_list)
plt.xlabel('a0')
plt.show()
```

```python
plt.hist(a1_list)
plt.xlabel('a1')
plt.show()
```

```python
plt.hist(train_mqa_list)
plt.xlabel('Train MQA')
plt.show()
```

```python
plt.hist(test_mqa_list)
plt.xlabel('Test MQA')
plt.show()
```
